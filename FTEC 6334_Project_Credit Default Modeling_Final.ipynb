# Data Preparation

### 1. Missing Values Imputation
Replace missing values with the most frequent values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = "most_frequent")
df = pd.DataFrame(imputer.fit_transform(df), columns = df.columns)
```

### 2. Encoding

```python
# Encoding for "Sex": Female as 1
Sex = pd.get_dummies(df.SEX)
df ["Female"] = Sex.Female

#Encoding for "Education"
Edu = pd.get_dummies(df.EDUCATION)
Edu.columns = [c.replace(' ', '_') for c in Edu.columns]
df ["Graduate"] = Edu.Graduate
df ["University"] = Edu.University
df ["High School"] = Edu.High_School

#Encoding for "Marriage"
Marr = pd.get_dummies(df.MARRIAGE)
df ["Single"] = Marr.Single
df ["Married"] = Marr.Married

df = df.drop(["SEX", "EDUCATION", "MARRIAGE"],axis=1)

df1=df.astype(float)
df1.head(10)
```

```python
np.transpose(df1.describe())
```

# Data Sampling

```python
Xs = df1.iloc[:, 2:28].values
Y = df1.iloc[:, 1].values
```

```python
#Stratified
from sklearn.model_selection import train_test_split
Xs_train, Xs_test, Y_train, Y_test = train_test_split(Xs, Y, test_size=0.2, random_state=0, stratify=Y)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import RFECV
```

##### Feature Selection: Recursive Feature Elimination with Cross-Validation

#### Logistic Model

```python
from sklearn.linear_model import LogisticRegression

#Feature Selection
rfe_classifier = RFECV(estimator=LogisticRegression())
model_classifier = LogisticRegression(solver = 'lbfgs')

# Pipeline to transform the data: rfe for the feature selection step and the model thatâ€™ll be used in the next step
pipe_classifier = Pipeline([('Feature Selection', rfe_classifier), ('Model', model_classifier)])
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=2)
n_scores = cross_val_score(pipe_classifier, Xs_train, Y_train, scoring='roc_auc', cv=cv, n_jobs=-1)
np.mean(n_scores)

#Fit the pipe to dataset
pipe_classifier.fit(Xs_train, Y_train)
```

```python
print("Optimal number of features : %d" % rfe_classifier.n_features_)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfe_classifier.grid_scores_) + 1), rfe_classifier.grid_scores_)
plt.show()
```

```python
rfe_classifier_df=pd.DataFrame(rfe_classifier.ranking_,index=df1.iloc[:, 2:28].columns,
                               columns=['Rank'])
rfe_classifier_df['Relative Rank'] = rfe_classifier.support_
rfe_classifier_df.sort_values(by='Rank',ascending=True)
```

```python
y_pred = pipe_classifier.predict(Xs_test).astype(int)
test = pd.DataFrame(np.concatenate((y_pred.reshape(len(y_pred),1), Y_test.reshape(len(Y_test),1)),1), 
                    columns = ['Predicted', 'Actual'])

cm = confusion_matrix(Y_test, y_pred)
predicted_prob = pipe_classifier.predict_proba(Xs_test)
print(f'Confusion Matrix:\n{cm}')
print(f'Accuracy Score: {accuracy_score(Y_test, y_pred)}')
print(f'AUC Score: {roc_auc_score(Y_test, predicted_prob[:,1])}')
test.head()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
def build_NNCmodel (size, rate):
    NNC = Sequential()
    for i in range(len(size)):
        NNC.add(Dense(units=size[i],kernel_initializer='glorot_uniform',
                    activation = 'relu'))
        for j in range(len(rate)):
            NNC.add(Dropout(rate[j]))       #only use 90% connection in each node
    #add output layer    
    NNC.add(Dense(units=1,kernel_initializer='glorot_uniform',
                    activation = 'sigmoid'))
    #compile NN
    NNC.compile(optimizer="adam",loss='binary_crossentropy',metrics=['accuracy'])
    return NNC
```

```python
NNCmodel = KerasClassifier(build_fn=build_NNCmodel)
```

##### Grid Search

```python
# create a dictionary of hyper-parameters to optimize
size = [[26], [51,26], [51,51,26]]
rate = [[0.1], [0.2], [0.5]]
hyperparameters = {"size": size, 'batch_size':[32,64,128], 'epochs':[10,20], 'rate':rate}
grid_search = GridSearchCV(estimator = NNCmodel, param_grid = hyperparameters, scoring = 'roc_auc', cv=3)
grid_search = grid_search.fit(Xs_train_NN,Y_train)
```
